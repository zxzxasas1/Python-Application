```python
# 使用 selenium 的瀏覽器模擬功能，於 momo 購物網搜尋關鍵字並擷取商品資料

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()

driver.implicitly_wait(30)

driver.get("https://www.momoshop.com.tw/main/Main.jsp")

target = driver.find_element(By.ID,"keyword")

target.send_keys("耳機" + Keys.ENTER)

wait = WebDriverWait(driver,300)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"body > div > div.bt_2_layout.searchbox.searchListArea > div.searchPrdListArea.bookList > div > ul > li")))

content = driver.find_elements(By.CSS_SELECTOR,"body > div > div.bt_2_layout.searchbox.searchListArea > div.searchPrdListArea.bookList > div > ul > li")

print(len(content))

for item in content : 

    title = item.find_element(By.CSS_SELECTOR,"div.goodsUrl > div.prdNameTitle > h3").text

    print(title)
```

```python
# 使用 selenium 的瀏覽器模擬功能，於 pchome 搜尋關鍵字並擷取商品資料

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas

driver = webdriver.Chrome()

driver.implicitly_wait(30)

driver.get("https://24h.pchome.com.tw")

target = driver.find_element(By.CLASS_NAME,"c-search__input")

target.send_keys("耳機" + Keys.ENTER)

wait = WebDriverWait(driver,300)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"body > div.l-layout.l-layout--rwd > main > div.container.container--lg > div > div > section:nth-child(2) > div > div > section > div > div > div.c-boxGrid > div.c-boxGrid__body.c-boxGrid__body--topBorder > div > div > ul:nth-child(1) > li")))

content = driver.find_elements(By.CSS_SELECTOR,"body > div.l-layout.l-layout--rwd > main > div.container.container--lg > div > div > section:nth-child(2) > div > div > section > div > div > div.c-boxGrid > div.c-boxGrid__body.c-boxGrid__body--topBorder > div > div > ul:nth-child(1) > li")

print(len(content))

data = []

for item in content : 

    title = item.find_element(By.CLASS_NAME,"c-prodInfoV2__title").text

    data.append(title)

    print(title)


result = pandas.DataFrame(data)

result.to_csv("output.csv",index=False)


```
