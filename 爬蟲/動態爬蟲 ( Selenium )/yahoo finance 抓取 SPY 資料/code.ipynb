```python
# 利用 selenium 的 sendkeys 和 click 來抓取 yahoo finance 上的 S&P 500 ETF 每日價格資料

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import pandas

driver = webdriver.Chrome()

driver.get("https://finance.yahoo.com")

target = driver.find_element(By.ID,"ybar-sbq")

target.send_keys("SPY" + Keys.RETURN)

wait = WebDriverWait(driver,300)
wait.until(expected_conditions.presence_of_element_located((By.LINK_TEXT,"Historical Data")))

target_2 = driver.find_element(By.LINK_TEXT,"Historical Data")

target_2.click()

wait_2 = WebDriverWait(driver,300)

table = wait_2.until(expected_conditions.visibility_of_element_located((By.TAG_NAME,"table")))

tr = table.find_elements(By.TAG_NAME,"tr")

data = []

for item in tr : 

    row = item.text

    data.append(row)

result = pandas.DataFrame(data)

result.to_csv("yahoo_finance_output.csv",index=False)
```

```python
# 較複雜版本，區分表格欄位和表格資料，但很常在 .text 轉換部分出錯，且表格欄位名稱抓不到
# 後來改用 visibility of element located 來抓取 table 修正

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import pandas

driver = webdriver.Chrome()

driver.get("https://finance.yahoo.com")

target = driver.find_element(By.ID,"ybar-sbq")

target.send_keys("SPY" + Keys.RETURN)

wait = WebDriverWait(driver,10)
wait.until(expected_conditions.presence_of_element_located((By.LINK_TEXT,"Historical Data")))

target_2 = driver.find_element(By.LINK_TEXT,"Historical Data")

target_2.click()

wait_2 = WebDriverWait(driver,10)

table = wait_2.until(expected_conditions.visibility_of_element_located((By.TAG_NAME,"table")))

tr_rows = table.find_elements(By.TAG_NAME,"tr")

data = []

for row in tr_rows : 

    th_rows = row.find_elements(By.TAG_NAME,"th")

    if th_rows : 

        th_data = []

        for th_cell in th_rows : 

            data_th_cell = th_cell.text

            th_data.append(data_th_cell)

        data.append(th_data)

    td_rows = row.find_elements(By.TAG_NAME,"td")

    if td_rows : 

        td_data = []

        for td_cell in td_rows : 
            
            data_td_cell = td_cell.text

            td_data.append(data_td_cell)

        data.append(td_data)

result = pandas.DataFrame(data)

result.to_csv("yahoo_finance_output_2.csv",index=False)
```
