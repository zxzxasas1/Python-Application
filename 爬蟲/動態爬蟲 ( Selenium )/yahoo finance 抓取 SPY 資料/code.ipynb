```python
# 利用 selenium 的 sendkeys 和 click 來抓取 yahoo finance 上的 S&P 500 ETF 每日價格資料

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import pandas

driver = webdriver.Chrome()

driver.get("https://finance.yahoo.com")

target = driver.find_element(By.ID,"ybar-sbq")

target.send_keys("SPY" + Keys.RETURN)

wait = WebDriverWait(driver,300)
wait.until(expected_conditions.presence_of_element_located((By.LINK_TEXT,"Historical Data")))

target_2 = driver.find_element(By.LINK_TEXT,"Historical Data")

target_2.click()

wait_2 = WebDriverWait(driver,300)
wait_2.until(expected_conditions.presence_of_element_located((By.TAG_NAME,"table")))

table = driver.find_element(By.TAG_NAME,"table")

tr = table.find_elements(By.TAG_NAME,"tr")

data = []

for item in tr : 

    row = item.text

    data.append(row)

result = pandas.DataFrame(data)

result.to_csv("yahoo_finance_output.csv",index=False)
```
