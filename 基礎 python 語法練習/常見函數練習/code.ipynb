```python
# 內建常見的函數練習 ( 多數是 string 型別的方法，因此只適用於字串 )



# 1. join 和 split 函數


# a. join : 

# join 是把 iterable 的物件，像是 list , tuple 等合併成一個字串的函數

# 通常是前面加上各元素要連接的文字或符號，例如 \t (tab) 或 \n (換行) 來產生不同效果

a = ["1","2","3"]

b = "\t".join(a)

c = "\n".join(a)

print(b)

print(c)

# 因為是 string 的函數，所以元素一定都要是字串才可以使用，否則要先用 str() 來轉換

num_a = [1,2,3]

d = ",".join([str(i) for i in num_a])

print(d)


# b. split : 

# split 就是 join 的相反，把字串依照條件拆開並產生 list

file_name = "20250801.csv"

file_name_split = file_name.split(".")

print(file_name_split)



# 2. splitlines 和 strip 函數


# a. splitlines : 

# 基本上就是 split + \n , 就是 split 的多行版本，把多行文字分開變成 list

lines = "a\n \nb\nc"

split_lines = lines.splitlines()

print(split_lines)


# b. strip : 

# 可以把字串前後的空白、tab 空白 \t、換行符號 \n 給移除，不影響中間字串

text = "\t meow \n"

print(text)

strip_text = text.strip()

print(strip_text)

# 也可以移除特定符號，參數指定即可 ( 如果有不同符號，就多 strip 幾次即可 )

text_2 = "$meow$"

print(text_2)

strip_text_2 = text_2.strip("$")

print(strip_text_2)


# c. 使用範例 : join , strip ,  splitlines 一起把 csv 資料內不要的資料行移除並合併回來

# 通常 csv 資料就是一列一列的字串，沒有上引號來包住每個欄位的資料，如下面 csv_data 印出來那樣

csv_data = "此表為每日價格\n日期,名稱,價格\n20250801,a,1\n\n20250802,b,2"

print(csv_data)

# 第一行為說明，所以需要移除，中間有空白行，也需要移除

split_csv_data = csv_data.splitlines()

join_csv_data = "\n".join(split_csv_data[1:])

# 移除掉開頭說明了，但因為還有空白行，所以要再用 if + strip 來確認是不是空白行，如果 strip 完有東西就留下來

split_csv_data_2 = join_csv_data.splitlines()

data_rows = [item for item in split_csv_data_2 if item.strip()]

join_csv_data_2 = "\n".join(data_rows)

print(join_csv_data_2)



# 3. replace 函數

# 就是以前 notepad 和 word 的搜尋替換，可以指定搜尋某個字元並換成另一個字元

# 在證交所的 csv 裡面，常常會有欄位多出 = ,這個時候就可以用 replace 拿掉

string_data = "日期,證券代號,價格\n20250801,=0001,20\n20250802,=0001,20.3"

new_data = string_data.replace("=","")

print(new_data)



# 4. endswith 函數

# 判斷字尾是否符合特定條件並產生布林值的函數，常用來判斷是否為某個檔案類型

file_list = ["aaa.txt","bbb.txt","ccc.csv"]

csv_file = [file for file in file_list if file.endswith(".csv")]

print(csv_file)



# 5. count 函數

# 計算字串內出現某個字元的次數，常常用來判斷 csv 檔案文字每一行的逗號有幾個，是不是有缺並移除資料不足的行

csv_text = "日期,證券代號,價格\n20250801,a,20\n20250801,a"

split_text = csv_text.splitlines()

new_csv_text = [text for text in split_text if text.count(",") > 1]

join_text = "\n".join(new_csv_text)

print(join_text)



```
