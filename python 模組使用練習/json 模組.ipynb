```python
# 練習 json 模組

import json



# 1. 專門接 API 的 json 字串的方法 : json.loads() 和 json.dumps() ( 字尾有 s 表示 string，可以這樣記 )


# a. json.loads() : 就是把 json 格式字串 loads 進來，解析變成 python dictionary 物件來操作

# 基本上就是字典樣子的字串

json_string = '{ "name" : "Tom", "age" : 20, "address" : "aaa"}'

# 把 json 字串用 json.loads() 來解析轉換成 dictionary

dictionary = json.loads(json_string)

print(dictionary)


# b. json.dumps() : 把製作完成的字典物件轉換成 json 字串倒出去 ( 序列化 , serializarion )

# 只有基本的 string , number 和 list 等可以轉換，其他不行，像是 datetime , set 等等不行

json_string_2 = json.dumps(dictionary)

print(json_string_2)


# c. json list 也可以用上面方式操作

json_list = '[{ "name" : "Tom", "age" : 20, "address" : "aaa"} , { "name" : "cat", "age" : 25, "address" : "bbb"}]'

dictionary_2 = json.loads(json_list)

print(dictionary_2)

json_string_3 = json.dumps(dictionary_2)

print(json_string_3)



# 2. 專門讀寫 json 檔案的方法 : json.load() 和 json.dump() ( 標準名字，所以就是給一般作法用，讀寫檔案 )


# a. json.load() : 把 json 檔案 load 進來，解析變成 python dictionary 物件來操作

# 因為是檔案讀寫，所以要用 open() 來操作

with open("data & output/json_example.json","r",encoding="utf-8-sig") as file : 

    dictionary_3 = json.load(file)

print(dictionary_3)


# b. json.dump() : 把 dictionary 內容寫進 json 檔案倒出去

dictionary_4 = {

    "name" : "Tom",
    "age" : 20,
    "address" : "aaa"

}

# 第一個參數放要寫入的資料，第二個放寫入目標檔案，indent 是縮排，比較好看，如果有中文，要用 ensure_ascii = False

with open("data & output/json_example.json","w",encoding="utf-8-sig") as file : 

    json.dump(dictionary_4,file,indent=2)


```
