```python
# 1. 練習製作折線圖

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
attendance_1 = [95.2,93.8,96.1,94.5,97.3,95.8]
attendance_2 = [85.8,87.3,84.5,86.1,83.8,85.2]


# a. 總是先從 plt.figure 開始，建立圖表物件，figsize 第一個參數是 x 軸長度 , 第二個是 y 軸長度 ( 不要加也是可以畫 , 會自動用預設長寬 )

plt.figure(figsize=(10,5))


# b. 引入 x , y 軸資料 , 想要多筆資料在同一張圖上就做多次

# ( 其他可調參數 : marker 是點的樣式，markersize 是點的大小，linewidth 是線的粗度 )

plt.plot(months,attendance_1,label="出席率_1",marker="o",markersize=6,linewidth=2)
plt.plot(months,attendance_2,label="出席率_2",marker="s",markersize=6,linewidth=2)


# c. 對每個資料點加上註記 , 用迴圈和 enumerate 以及 annotate() 來做 , 有多筆資料就一樣做多次

# ( annotate 第一個參數是要標記的值(記得要是字串) , 第二個是 tuple 的 xy 位置 (因為 x 軸這邊是字串,所以用索引 i 即可 , y 就是y軸的值來表達位置) ,  

#   剩下就是距離的計算基準單位 , xy 的位置偏移量 , 對齊方式 , 以及顯示字體大小 )

for i , item in enumerate(attendance_1) : 

    plt.annotate(f"{item}" , (i,item) , textcoords="offset points" , xytext=(0,-20) , ha="center", fontsize=12)

for i , item in enumerate(attendance_2) : 

    plt.annotate(f"{item}" , (i,item) , textcoords="offset points" , xytext=(0,10) , ha="center", fontsize=12)


# d. 對 x , y 軸標籤以及表頭命名

# ( 其他可調參數 : fontsize 是字體大小，fontweight 是字體粗細樣式 )

plt.title("出席統計",fontsize=16,fontweight="bold")
plt.xlabel("月份",fontsize=12)
plt.ylabel("出席率",fontsize=12)


# e. 對圖表做其他微調 , xticks 是 x 軸刻度呈現 , grid 是隔線 , legend 是標記

plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()


# f. 最後就可以用 show 畫出圖

plt.show()
```

```python
# 2. 練習製作分布圖

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
attendance_1 = [95.2,93.8,96.1,94.5,97.3,95.8]
attendance_2 = [85.8,87.3,84.5,86.1,83.8,85.2]

plt.figure(figsize=(10,5))


# 不一樣的地方就是把 plt.plot 改成 plt.scatter 而已，注意沒有 markersize 的調整參數 , 也沒有線的調整參數

plt.scatter(months,attendance_1,label="出席率_1",marker="o")
plt.scatter(months,attendance_2,label="出席率_2",marker="s")


for i , item in enumerate(attendance_1) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,-20),ha="center",fontsize=12)

for i , item in enumerate(attendance_2) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)


plt.title("出席統計",fontsize=16,fontweight="bold")
plt.xlabel("月份",fontsize=12)
plt.ylabel("出席率",fontsize=12)

plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()

plt.show()
```

```python
# 3. 垂直直方圖

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
product_A = [55,45,40,45,55,45]
product_B = [30,25,30,35,30,30]

plt.figure(figsize=(10,5))


# 把 plt.plot 改成 plt.bar

plt.bar(months,product_A,label="產品A")
plt.bar(months,product_B,label="產品B")


for i , item in enumerate(product_A) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)

for i , item in enumerate(product_B) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)


plt.title("產品銷售額統計",fontsize=16,fontweight="bold")
plt.xlabel("月份",fontsize=12)
plt.ylabel("銷售額",fontsize=12)


# 要注意因為直方圖本來長條就有對到 x 軸 , 所以 grid 可以省略 x 軸格線 , 只留 y 軸格線

plt.grid(axis="y")


plt.xticks(rotation=45)
plt.legend()

plt.show()
```

```python
# 4. 水平直方圖

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
product_A = [55,45,40,45,55,45]
product_B = [30,25,30,35,30,30]

plt.figure(figsize=(10,5))


# 把 plt.plot 改成 plt.barh , 要注意水平直方圖的 x 軸 和 y 軸是相反的 , x 軸放數值 , y 軸放類別 ,

# 但是函數本身已經自己有調整過 , 所以還是照原本的 x , y 軸放第一個和第二個參數就可以

plt.barh(months,product_A,label="產品A")
plt.barh(months,product_B,label="產品B")


# annotate 也要自己互換 x , y 的資料位置 , 才能對到正確的位置

for i , item in enumerate(product_A) : 

    plt.annotate(f"{item}",(item,i),textcoords="offset points",xytext=(-20,-5),ha="center",fontsize=12)

for i , item in enumerate(product_B) : 

    plt.annotate(f"{item}",(item,i),textcoords="offset points",xytext=(-20,-5),ha="center",fontsize=12)


# x , y 軸的 label , ticks 和 grid 沒有自動互換 , 所以要自己換過來

plt.title("產品銷售額統計",fontsize=16,fontweight="bold")
plt.xlabel("銷售額",fontsize=12)
plt.ylabel("月份",fontsize=12)

plt.yticks(rotation=45)
plt.grid(axis="x")
plt.legend()

plt.show()
```

```python
# 5. 長條圖

import numpy
import matplotlib.pyplot as plt

group_A = numpy.random.normal(0,3,10000)
group_B = numpy.random.normal(12,5,10000)

plt.figure(figsize=(10,5))


# 由 plt.plot 改成 plt.hist , 第一個參數放 y 軸數值 , x 軸為 bins 決定有幾組

plt.hist(group_A,bins=100,label="Group_A")
plt.hist(group_B,bins=100,label="Group_B")


plt.title("Normal Distribution",fontsize=16,fontweight="bold")
plt.xlabel("numbers",fontsize=12)
plt.ylabel("times",fontsize=12)

plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.show()
```

```python
# 6. 圓餅圖

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
product = [55,45,40,45,55,45]
color_set = ["#eb7412","#e3e616","#2ae219","#1ed8b0","#20a6db","#1e70da"]

plt.figure(figsize=(10,5))


# 由 plt.plot 改成 plt.pie , 和長條圖一樣沒有 x 軸 y 軸的概念 , 而是分組 , 第一個參數放 y 軸數值 , 第二個放 x 軸 類別 (labels)

# 要注意要設定 autopct 來顯示每一組的百分比 , shadow 可以產生陰影效果 , startangle 可以調整起始位置角度 , textprops 來調整字體大小

# color 可以打 #ffffff 或 #000000 先找到白色或黑色並產生調整按鈕，再調整到自己洗要的顏色

plt.pie(product,labels=months,autopct="%1.1f%%",shadow=True,startangle=90,textprops={"fontsize" : 12},colors=color_set)


# 沒有 x 哲和 y 軸 , 所以不用 label , 但要 axis 來調整圖片 x 和 y 軸的比例，通常都設定相等

plt.title("銷售額統計",fontsize=16,fontweight="bold")
plt.axis("equal")


# 也不用 ticks 和 grid

plt.legend()

plt.show()

```
