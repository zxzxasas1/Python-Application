```python
# 1. 

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1","2","3","4","5","6"]
attendance_1 = [95.2,93.8,96.1,94.5,97.3,95.8]
attendance_2 = [85.8,87.3,84.5,86.1,83.8,85.2]


# a.  plt.figure figsize  x  ,  y  (  ,  )

plt.figure(figsize=(10,5))


# b.  x , y  , 

# (  : marker markersize linewidth  )

plt.plot(months,attendance_1,label="_1",marker="o",markersize=6,linewidth=2)
plt.plot(months,attendance_2,label="_2",marker="s",markersize=6,linewidth=2)


# c.  ,  enumerate  annotate()  , 

# ( annotate () ,  tuple  xy  ( x , i  , y y) ,  

#    , xy  ,  ,  )

for i , item in enumerate(attendance_1) : 

    plt.annotate(f"{item}" , (i,item) , textcoords="offset points" , xytext=(0,-20) , ha="center", fontsize=12)

for i , item in enumerate(attendance_2) : 

    plt.annotate(f"{item}" , (i,item) , textcoords="offset points" , xytext=(0,10) , ha="center", fontsize=12)


# d.  x , y 

# (  : fontsize fontweight  )

plt.title("",fontsize=16,fontweight="bold")
plt.xlabel("",fontsize=12)
plt.ylabel("",fontsize=12)


# e.  , xticks  x  , grid  , legend 

plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()


# f.  show 

plt.show()
```

```python
# 2. 

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1","2","3","4","5","6"]
attendance_1 = [95.2,93.8,96.1,94.5,97.3,95.8]
attendance_2 = [85.8,87.3,84.5,86.1,83.8,85.2]

plt.figure(figsize=(10,5))


#  plt.plot  plt.scatter  markersize  , 

plt.scatter(months,attendance_1,label="_1",marker="o")
plt.scatter(months,attendance_2,label="_2",marker="s")


for i , item in enumerate(attendance_1) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,-20),ha="center",fontsize=12)

for i , item in enumerate(attendance_2) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)


plt.title("",fontsize=16,fontweight="bold")
plt.xlabel("",fontsize=12)
plt.ylabel("",fontsize=12)

plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.legend()

plt.show()
```

```python
# 3. 

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1","2","3","4","5","6"]
product_A = [55,45,40,45,55,45]
product_B = [30,25,30,35,30,30]

plt.figure(figsize=(10,5))


#  plt.plot  plt.bar

plt.bar(months,product_A,label="A")
plt.bar(months,product_B,label="B")


for i , item in enumerate(product_A) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)

for i , item in enumerate(product_B) : 

    plt.annotate(f"{item}",(i,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)


plt.title("",fontsize=16,fontweight="bold")
plt.xlabel("",fontsize=12)
plt.ylabel("",fontsize=12)


#  x  ,  grid  x  ,  y 

plt.grid(axis="y")


plt.xticks(rotation=45)
plt.legend()

plt.show()
```

```python
# 4. 

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1","2","3","4","5","6"]
product_A = [55,45,40,45,55,45]
product_B = [30,25,30,35,30,30]

plt.figure(figsize=(10,5))


#  plt.plot  plt.barh ,  x   y  , x  , y  ,

#  ,  x , y 

plt.barh(months,product_A,label="A")
plt.barh(months,product_B,label="B")


# annotate  x , y  , 

for i , item in enumerate(product_A) : 

    plt.annotate(f"{item}",(item,i),textcoords="offset points",xytext=(-20,-5),ha="center",fontsize=12)

for i , item in enumerate(product_B) : 

    plt.annotate(f"{item}",(item,i),textcoords="offset points",xytext=(-20,-5),ha="center",fontsize=12)


# x , y  label , ticks  grid  , 

plt.title("",fontsize=16,fontweight="bold")
plt.xlabel("",fontsize=12)
plt.ylabel("",fontsize=12)

plt.yticks(rotation=45)
plt.grid(axis="x")
plt.legend()

plt.show()
```

```python
# 5. 

import numpy
import matplotlib.pyplot as plt

group_A = numpy.random.normal(0,3,10000)
group_B = numpy.random.normal(12,5,10000)

plt.figure(figsize=(10,5))


#  plt.plot  plt.hist ,  y  , x  bins 

plt.hist(group_A,bins=100,label="Group_A")
plt.hist(group_B,bins=100,label="Group_B")


plt.title("Normal Distribution",fontsize=16,fontweight="bold")
plt.xlabel("numbers",fontsize=12)
plt.ylabel("times",fontsize=12)

plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.show()
```

```python
# 6. 

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1","2","3","4","5","6"]
product = [55,45,40,45,55,45]
color_set = ["#eb7412","#e3e616","#2ae219","#1ed8b0","#20a6db","#1e70da"]

plt.figure(figsize=(10,5))


#  plt.plot  plt.pie ,  x  y  ,  ,  y  ,  x   (labels)

#  autopct  , shadow  , startangle  , textprops 

# color  #ffffff  #000000 

plt.pie(product,labels=months,autopct="%1.1f%%",shadow=True,startangle=90,textprops={"fontsize" : 12},colors=color_set)


#  x  y  ,  label ,  axis  x  y 

plt.title("",fontsize=16,fontweight="bold")
plt.axis("equal")


#  ticks  grid

plt.legend()

plt.show()

```
