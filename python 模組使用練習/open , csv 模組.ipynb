```python
# 內建開啟檔案函數 open 的使用 

# open : 路徑 + 動作 ( r , rb , w ) + 解碼

# 路徑 : 如果從系統資料夾路徑 copy 記得要把 \ 改成 / 才能讀到，不然就是檔案直接 copy 路徑

# 動作 : 還有 a ( append ) 和 x ( create )

import csv

import pandas

# .read() 所有檔案都能讀取，直接讀取整個檔案，但輸出結構也簡單，單純為一整個字串 

# ( readline : 讀取單行而已，輸出字串，如果要讀取多行就要一直 readline ; readlines : 讀取所有行，輸出 list )

with open("D:/Python/python 模組使用練習/data & output/20250802.csv", "r") as file : 

    read_file = file.read()

    print(read_file)

# csv.reader() 能逐行讀取 csv 檔，但讀取出來是 csv.reader 物件，是一個 iterator，要用 for 讀取元素或 list 轉換

# csv.reader 可以設定引數來接收不同分隔符號的 csv 檔

with open("D:/Python/python 模組使用練習/data & output/20250802.csv", "r") as file : 

    csv_file = csv.reader(file)

    for row in csv_file : 

        print(row)

# pandas.read_csv() 能完整讀取並轉成 dataframe 分析，是比較推薦的做法

with open("D:/Python/python 模組使用練習/data & output/20250802.csv", "r") as file : 

    pandas_file = pandas.read_csv(file)

    print(pandas_file)
    


data = "123"

data_1 = [1,2,3]

data_2 = [

    ["name","age","address"],

    ["Tom",20,"aaa"],

    ["Cat",25,"bbb"],

    ["Meow",30,"ccc"]

]

data_3 = [
    
    {

        "name" : "Tom",
        "age" : 20,
        "address" : "aaa"

    },

    {

        "name" : "Cat",
        "age" : 25,
        "address" : "bbb"

    },

    {

        "name" : "Meow",
        "age" : 30,
        "address" : "ccc"

    }

]

# 內建函數 open 改成作為寫入的使用，同時亦可建立檔案

# 用內建的函數 write 只能寫入字串 ( 重複多次可以一直接下去 ) , writelines 是寫入 list 但裡面元素也要是字串 

# ( read / write : 讀寫都是一整個字串、readlines / writelines : 讀寫都是字串 list )

with open("D:/Python/python 模組使用練習/data & output/write.csv", "w") as file : 

    file.write(data)

# writelines 是寫入 list，設計是為了每行寫入，但不會自己分行，要自己加 \n ( 這邊 data 是字串，但字串也可以當 list，元素就是字元 )

with open("D:/Python/python 模組使用練習/data & output/write_2.csv", "w") as file : 

    paragraph_data = [item + "\n" for item in data]

    file.writelines(paragraph_data)

# list 裡面如果是非字串，要先轉成字串才能用 writelines 寫入 

# ( 因為沒有加 \n 分行，所以結果會是一整個字串，和 write 結果一樣 )

with open("D:/Python/python 模組使用練習/data & output/write_3.csv", "w") as file : 

    string_data = [str(item) for item in data_1]

    file.writelines(string_data)

# 用 csv.writer 可以把 csv 結構的資料輸出成 csv 檔案，像是 list of list , 直接用 writerows 多行寫入即可，不用 writerow

# 要記得用 newline = "" 避免 csv.writer 產生空行

with open("D:/Python/python 模組使用練習/data & output/write_4.csv","w", newline="") as file :

    writer = csv.writer(file)

    writer.writerows(data_2)

# list of list 的 csv 也可以不用 csv.writer , 單純用 for + write + \n 來調整成每行顯示

# ( 如果用 writelines 就要在 for 過程把資料轉回成 list 才能寫入 )

# 要用 for 來讀取 list of list 的每一行 list，然後再用 join 函數把 list 裡面元素合併並用逗號或 \t 分開    

with open("D:/Python/python 模組使用練習/data & output/write_5.csv","w") as file :

    for row in data_2 :

        # 記得要把元素都轉換成字串才能 join，因為是字串型別的方法

        new_row = [str(item) for item in row]

        join_row = ",".join(new_row) + "\n"

        file.write(join_row)

# list of dictionary 可以用 csv 的 csv.DictWriter 來寫入

# 同樣要記得用 newline = "" 避免 csv.DictWriter 產生空行

with open("D:/Python/python 模組使用練習/data & output/write_6.csv","w", newline="") as file :

    # 要先找到資料欄位的 list

    titles = list(data_3[0].keys())

    # 就可以寫入

    writer_2 = csv.DictWriter(file , fieldnames=titles)

    writer_2.writeheader

    writer_2.writerows(data_3)

#  pandas 的 .to_csv() 方法來寫入輸出 csv 檔 , 要先轉成 dataframe 才能用 , 就不用 open 函數

    dataframe = pandas.DataFrame(data_2)

    dataframe.to_csv("data & output/write_7.csv",index=False)
```
