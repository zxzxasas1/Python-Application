```python
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
attendance_1 = [95.2,93.8,96.1,94.5,97.3,95.8]
attendance_2 = [85.8,87.3,84.5,86.1,83.8,85.2]

plt.figure(figsize=(10,5))

plt.plot(months,attendance_1,marker="o",label='出席率1',linewidth=2,markersize=6)
plt.plot(months,attendance_2,marker="s",label='出席率2',linewidth=2,markersize=6)

plt.title("出勤統計",fontsize=16,fontweight="bold")
plt.xlabel("月份",fontsize=12)
plt.ylabel("出勤率",fontsize=12)



for n , item in enumerate(attendance_1) : 

    plt.annotate(f"{item}",(n,item),textcoords="offset points",xytext=(0,-20),ha="center",fontsize=12)

for n , item in enumerate(attendance_2) : 

    plt.annotate(f"{item}",(n,item),textcoords="offset points",xytext=(0,10),ha="center",fontsize=12)


plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

months = ["1月","2月","3月","4月","5月","6月"]
product_A = [55,45,40,45,55,45]
product_B = [30,25,30,35,30,30]

plt.figure(figsize=(10,5))

plt.barh(months,product_A,label='產品A')
plt.barh(months,product_B,label='產品B')

plt.title("產品銷售統計",fontsize=16,fontweight="bold")
plt.xlabel("產品銷售",fontsize=12)
plt.ylabel("月份",fontsize=12)

for n , item in enumerate(product_A) : 

    plt.annotate(f"{item}",(item,n),textcoords="offset points",xytext=(-20,0),ha="left",fontsize=12)

for n , item in enumerate(product_B) : 

    plt.annotate(f"{item}",(item,n),textcoords="offset points",xytext=(-20,0),ha="left",fontsize=12)


plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.legend()
plt.show()
```

```python
import numpy
import matplotlib

random_A = numpy.random.normal(loc=0,scale=3,size=10000)
random_B = numpy.random.normal(loc=12,scale=5,size=10000)

plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei" , "Simhei"]
plt.rcParams["axes.unicode_minus"] = False

plt.figure(figsize=(10,5))

plt.hist(random_A,bins=100,label="group A")
plt.hist(random_B,bins=100,label="group B")

plt.title("常態分配",fontweight="bold")
plt.xlabel("數值")
plt.ylabel("數量")

plt.grid(axis="y")
plt.legend()
plt.show()
```
